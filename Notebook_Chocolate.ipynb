```python
def distribution(data, title, x_label, y_label, figsize=(10,5)):
    plt.figure(figsize=figsize)
    plt.style.use('seaborn-pastel')
    density = sns.barplot(x=data.index, y=data.values)
    title = plt.title(title, fontdict={'fontsize': 20})
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

rating_REF_dist = df_train.groupby('Rating')['REF'].mean()

distribution(data=rating_REF_dist,
             title='Mean REF in rating',
             x_label='Rating',
             y_label='REF')
```

```python
print(f"REF (min, max) = {df_train['REF'].min()}, {df_train['REF'].max()}")
REF_bins = np.arange(0, 2100, 100)
REF_bin_centers = [(REF_bins[i+1] + REF_bins[i]) / 2 for i in range(REF_bins.shape[0]-1)]
print(REF_bin_centers)
```

```python
def mean_rating_in_bins(data, col, bin_edges):
    bin_values = [
        data[(data[col] > bin_edges[i])
             & (data[col] <= bin_edges[i + 1])]['Rating'].mean()
        for i in range(len(bin_edges) - 1)
    ]
    return bin_values
```

```python
plt.plot(REF_bin_centers, mean_rating_in_bins(df_train, 'REF', REF_bins), 'o-r')
plt.xlabel('REF')
plt.ylabel('Rating mean')
plt.show()
```

```python
def length_bin(s: str, min_val=0, max_val=50, step=10) -> int:
    n = len(s)
    bins = np.arange(min_val, max_val+step, step)
    for i in range(bins.shape[0]):
        if n > bins[i] and n <= bins[i+1]:
            return i

df_train_copy['SBO length'] = df_train_copy['Specific Bean Origin'].apply(lambda x: length_bin(x))
display(df_train_copy.groupby('SBO length').agg({'Rating': 'mean', 'SBO length': 'count'}))
df_train_copy[['SBO length', 'Rating']].corr()
```

```python
# Most probable bean origins
top_bean_types = y.value_counts()[:7].index
top_bean_types_mask = y.apply(lambda x: x in top_bean_types)

# train on top 10 Broad Bean Origins only
X_top = X[top_bean_types_mask]
y_top = y[top_bean_types_mask]

X_train, X_val, y_train, y_val = train_test_split(X_top, y_top, test_size=0.2)

categorical_features_indices = np.where(X_top.dtypes == object)[0]
```

```python
# CatBoost with default hyperparameters

model = CatBoostClassifier()
#auto_class_weights='Balanced',
#loss_function='MultiClass'

model.fit(X_train, y_train,
          cat_features=categorical_features_indices,
          logging_level='Silent')

pred = model.predict(X_val)

label_accuracy(y_val, pred)
```

```python
def objective(trial):

    param = {
        "n_estimators": trial.suggest_int("n_estimators", 10, 1000),
        "max_depth": trial.suggest_int("max_depth", 1, 12),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.1, step=0.001)
    }

    estimator = CatBoostClassifier(**param,
                                   cat_features=categorical_features_indices,
                                   verbose=False)

    estimator.fit(X_train, y_train)
    pred = estimator.predict(X_val)

    return label_accuracy(y_val, pred)

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10, n_jobs=-1)
print(study.best_trial)
```

```python
print(study.best_trial.params)
```

```python
model = CatBoostClassifier(**study.best_params,
                           cat_features=categorical_features_indices,
                           logging_level='Silent')

model.fit(X_train, y_train)
pred = model.predict(X_val)

label_accuracy(y_val, pred)
```
